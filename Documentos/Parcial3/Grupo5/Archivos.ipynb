```python
import pandas as pd
import matplotlib.pylab as plt
import mpl_toolkits.mplot3d.axes3d as p3
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
#se lee el archivo csv y se convierte en un dataframe
df = pd.read_csv ("tubo.csv")
```

```python
df1 = pd.read_csv ("cuerda.csv")
#print(df1)

x_tubo=np.linspace(0,1,len(df.iloc[0]))
x_cuerda=np.linspace(0,3,len(df1.iloc[0]))
```

```python
fig1 = plt.figure(figsize=(8, 8))
ax1 = plt.axes(xlim=(0, 3), ylim=(-0.3, 0.3))
line1, = ax1.plot([], [])
ax1.grid()

ax1.set_title("cuerda pulsada")

def init1():
    line1.set_data([], [])
    return line1,

def animate_1(i):
    
    y = df1.iloc[i]
    line1.set_data(x_cuerda, y)
    return line1,

anim_cuerda = FuncAnimation(fig1, animate_1, init_func=init1,
                               frames=len(df1), interval=len(df1.iloc[0]),blit=True)


anim_cuerda.save('cuerda.gif', writer='imagemagick')
```

```python
fig_tubo = plt.figure(figsize=(8, 8))
ax_tubo = plt.axes(xlim=(0,1), ylim=(-1, 3))
line_tubo, = ax_tubo.plot([], [])
ax_tubo.grid()

ax_tubo.set_title("Presion en un tubo abierto")

def init_tubo():
    line_tubo.set_data([], [])
    return line_tubo,

def animate(i):
    
    y = df.iloc[i]
    line_tubo.set_data(x_tubo, y)
    return line_tubo,

anim_tubo = FuncAnimation(fig_tubo, animate, init_func=init_tubo,
                               frames=len(df), interval=len(df.iloc[0]),blit=True)


anim_tubo.save('tubo.gif', writer='imagemagick')
```
