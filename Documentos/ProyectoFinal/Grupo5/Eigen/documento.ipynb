```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
#se lee el archivo csv y se convierte en un dataframe
df = pd.read_csv ("datos.csv", header=None)
```

```python
df
```

```python
primer=np.array(df.loc[0])
```

```python
nuevo=primer[np.where(primer!='0)')]    
```

```python
nuevo
```

```python
for i in range(len(nuevo)):
    if(i!=1 and i!=len(nuevo)-1):
        nuevo[i]=float(nuevo[i][1:-1])
    if i==0:
        nuevo[i]*=10**6
    elif i==2:
        nuevo[i]*=10**(-8)
    elif i==3:
        nuevo[i]*=10**(-8)
    elif i==4:
        nuevo[i]*=10**(-7)
    elif i==5:
        nuevo[i]*=10**(-6)
    elif i==6:
        nuevo[i]*=10**(-6)
    elif i==7:
        nuevo[i]*=10**(-5)
    elif i==8:
        nuevo[i]*=10**(-5)
    elif i==313:
        nuevo[i]*=10**(-9)
    elif i==315:
        nuevo[i]*=10**(-6)
```

```python
nuevo
```

```python
x=np.linspace(-3.5*10**(-14),3.5*10**(-14),len(nuevo)-1)
```

```python
psi = nuevo[1:]
```

```python
plt.plot(x,psi**2)
plt.grid()
plt.savefig("600.jpg")
plt.show()
```
